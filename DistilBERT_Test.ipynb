```python
from transformers import pipeline

# Load pre-trained DistilBERT model for Sentiment Analysis
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Example texts
texts = [
    "I love using DistilBERT! It's so fast and efficient.",
    "This product is really bad. I regret buying it.",
    "The movie was okay, not too bad but not great either.haa"
]

# Perform sentiment analysis
results = sentiment_pipeline(texts)

# Print results
for text, result in zip(texts, results):
    print(f"Text: {text}\nSentiment: {result['label']} (Confidence: {result['score']:.4f})\n")
```

```python
from transformers import pipeline

# Load pre-trained DistilBERT model for NER (Name Recognition)
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", grouped_entities=True)

# Example text
text = ""

# Perform NER
entities = ner_pipeline(text)

# Print results
for entity in entities:
    print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Confidence: {entity['score']:.3f}")
```

```python
from transformers import pipeline

# Load Question Answering model
qa_pipeline = pipeline("question-answering")

# Context (passage)
context = "Albert Einstein was a theoretical physicist. He developed the theory of relativity."

# Question
question = "Who developed the theory of relativity?"

# Get answer
result = qa_pipeline(question=question, context=context)

# Print result
print(result)
```

```python
from transformers import pipeline

# Load Translation model
translator = pipeline("translation_en_to_fr")

# English text
text = "Artificial intelligence is transforming the world."

# Translate to French
translated_text = translator(text, max_length=50)

# Print translation
print(translated_text[0]['translation_text'])
```

```python
from transformers import pipeline

# Load Summarization model
summarizer = pipeline("summarization")

# Example long text
text = """Artificial Intelligence (AI) is transforming industries.
          AI-powered systems help businesses automate tasks, improve efficiency,
          and provide better customer experiences. The advancements in deep learning
          have led to significant progress in AI applications such as chatbots,
          self-driving cars, and medical diagnosis."""

# Generate summary
summary = summarizer(text, max_length=50, min_length=20, do_sample=False)

# Print summary
print(summary[0]['summary_text'])
```

```python
from transformers import pipeline

# Load Text Generation model (GPT-2)
text_generator = pipeline("text-generation", model="gpt2")

# Prompt text
prompt = "The future of artificial intelligence is"

# Generate text
generated_text = text_generator(prompt, max_length=50, num_return_sequences=1)

# Print generated text
print(generated_text[0]['generated_text'])
```

```python
from transformers import pipeline

# Load Zero-Shot Classification model
classifier = pipeline("zero-shot-classification")

# Example text
text = "I love watching movies with deep storytelling."

# Candidate labels
labels = ["Technology", "Entertainment", "Sports"]

# Perform classification
result = classifier(text, candidate_labels=labels)

# Print result
print(result)
```

## Question Answering (QA) to extract specific details from a resume

```python
# Load the Question Answering pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
resume_text = """
John Doe
Software Engineer
Email: johndoe@example.com
Phone: +1 123-456-7890
LinkedIn: linkedin.com/in/johndoe
GitHub: github.com/johndoe

Experience:
- Software Engineer at Google (2020 - Present)
- Intern at Microsoft (2019 - 2020)

Education:
- Master’s in Computer Science, Stanford University (2018 - 2020)
- Bachelor’s in Computer Science, MIT (2014 - 2018)

Skills:
- Python, Java, SQL
- Machine Learning, Deep Learning
- Cloud Computing (AWS, GCP)
"""
```

```python
# Function to extract information
def extract_info(question, context):
    result = qa_pipeline(question=question, context=context)
    return result["answer"]

# Questions
name = extract_info("What is the candidate's name?", resume_text)
email = extract_info("What is the candidate's email?", resume_text)
phone = extract_info("What is the candidate's phone number?", resume_text)
linkedin = extract_info("What is the candidate's LinkedIn profile link?", resume_text)
github = extract_info("What is the candidate's GitHub profile?", resume_text)
job_title = extract_info("What is the candidate's current job title?", resume_text)
company = extract_info("Where does the candidate currently work?", resume_text)
education = extract_info("What is the education level?", resume_text)
skills = extract_info("What are the candidate's skills?", resume_text)

# Print extracted details
print(f"Name: {name}")
print(f"Email: {email}")
print(f"Phone: {phone}")
print(f"LinkedIn: {linkedin}")
print(f"GitHub: {github}")
print(f"Job Title: {job_title}")
print(f"Current Company: {company}")
print(f"Highest Education: {education}")
print(f"Skills: {skills}")
```
